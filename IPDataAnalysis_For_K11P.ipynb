```python
ipldataset.batting_team.replace(to_replace='Chennai Super Kings', value='CSK', inplace=True)
ipldataset.bowling_team.replace(to_replace='Kings XI Punjab', value='K11P', inplace=True)

# Get top 5 CSK batsmen against K11P Bowlers
# Initially took top 5 batsmen against all Team, best bowlers data turned out to be less 

batsmen_of_CSK = ipldataset[(ipldataset.batting_team=='CSK') & (ipldataset.bowling_team=='K11P')] 
batsman_with_runs = batsmen_of_CSK[['batsman', 'batsman_runs'] ]
batsman_with_ball = batsmen_of_CSK[['batsman','ball'] ]
```

```python
batsman_with_ball = batsman_with_ball.groupby(['batsman']).count()
```

```python
batsman_with_runs = batsman_with_runs.groupby(['batsman']).sum()
```

```python
batsmen_runs_and_ball = pd.merge(batsman_with_ball, batsman_with_runs, on=['batsman'])
batsmen_runs_and_ball['run_per_ball'] = batsmen_runs_and_ball['batsman_runs'] / batsmen_runs_and_ball['ball']
```

```python
batsmen_runs_and_ball.sort_values(by='run_per_ball', ascending=False, inplace=True)
batsmen_runs_and_ball.index.array.to_numpy()
```

```python

plt.figure(figsize=(16,4))
sns.barplot(x=batsmen_runs_and_ball.index.array.to_numpy() , y='run_per_ball', data=batsmen_runs_and_ball)
plt.xticks(rotation = 90)
plt.title('Run per ball of each batsmen', fontsize = 18)
plt.show()
```

```python
top_5_batsmen_of_CSK = batsmen_runs_and_ball[0:5]
```

```python
top_5_batsmen_of_CSK
```

```python
# We need to find top 4 K11P bowlers against these top 5 CSk Batsmen

# Store bowling data of K11P and Batting as RCB

K11P_bowling_against_CSK = ipldataset[(ipldataset.bowling_team=='K11P') & (ipldataset.batting_team=='CSK')]
K11P_bowling_against_CSK



```

```python
K11P_bowler_with_top_5_CSK_player_dismissed = K11P_bowling_against_CSK[(K11P_bowling_against_CSK['player_dismissed'].isin(top_5_batsmen_of_CSK.index))]

K11P_bowler_with_top_5_CSK_player_dismissed
```
